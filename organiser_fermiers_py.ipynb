```python
import pandas as pd
import re

# Données des fermiers et codes
donnees_brutes = """
1:paluku muleha yoshiya
2: lumina Mustapha
3: Jacque  kavaka
4: kyester mumbere
5 :Jorgan  yoshiya
6: Roger vinyatsi
7: firmé mushabahaa
8 : sivitha Eva  kahambu
9: shabani balongo
10: Kambale triphin
11:  kisumb Samuel
12: mukulu mbusa
13:  muke phar
14: kiza kiza
15: bruno Pasteur
16: sivyolo denewade  kasereka
17 :  kisumba  charle
18: tj symphorien
19 : kambasu Jeremy
20 : aleksi palukuu
21 : kakule luvwa
22:  Ebenezer  syahingaa
23: kakungu
24: ngororio
25:Kasereka  kamutsere
26: bonet muhindo
27: savitos kakule kasomo
28: Elois sivanzire,
29: leandre
30: mbayithula
31: Jackson  kambale kavigha

1.KASEREKA MAHAMBA PIEUS:PH-MGN-MA-0200
2.KATOTA ELYZEE:PH-MGN-MU-0196
3.MAKA SELEO:PH-MGN-MU-0752
4.MBAFUMOJA AUGUSTIN:PH-MGN-MU-0765
5.PIKA MWENDO EZÉCHIEL :PH-MGN-MU-0876
6.MYAMBALIRE ANDRE:PH-MGN-MU-7501
7.MUTUPA JETHRO:PH-MGN-MU-0765
8.BONNE ANNÉE LYENDE:PH-MGN-MK-9866
9.KAMITATU OLIVIER :PH-MGN-MU-0782
10.KYOLOVIRE YOSHUA:PH-MGN-MU0332
11.MALONGA ADELARD :PH-MGN-MK-9853
12.KWALI KWA LAVA METHUSELAH:PH-MGN-MP-0156
13.MUVIRI ZAÏROIS :PH-MGN-MP-7824
14.MUKWETU JAPHET :PH-MGN-MA-3781
14.KAKULUKULU EMMANUEL :PH-MGN-TE-3679
15.SILWAMUGHUMA CALEB:CO-MGN-TE-0523
16.KASAI CHRISOSTOME:CO-MGN-TI-9036
17.RANGI JUPSON:CO-MGN-TI-9015
18.KOMBI NOEL:PH-MGN-KU-0328
19.VYAMBWERA DALMON:PH-MGN-KU-0754
20.BALIKWISHA KIBAMULYA MARCEL:PH-MGN-MU-0614

1KAKULE VALAMBIYI EISEE. PH-MTG-BU-N0
2KASEREKA WANZALABAMA ELISEE PH-MTG-MU-0160
3.KAMBALE FUNDIJOSKARE PH-MTG-MU-N252
4KASEREKA KALEMEKO PH-MTG-MU-N279
5KASEREKA MUYISA PH-MTG- MU- N026
6PALUKU KAMALOLWAKO PATRICK PH-MTG-MU-N0223
7PALUKU KATHOMDOLO PH-MTG-MU-N0245
8PALUKU WANZALABANA EPHREM PH-MTG-MU-N132
9PALUKU VILAMBIRYO BALELA PH-MTG-BU-N0289
10PALUKU MUKENGNGE CHABANTU PH-MTG-BU-N091
11 MUHINDO SAHANO CORNELLI PH-MTG-BU-N0304
12.MUHINDO NOWA5 MBOZE PH-MTG- BU-093
13.MUHINDO KAPAMBA BECKER PH-MTG-0414
14MBUSA MWALIKYA ZEPHANIE MTG-BU0166
15.KAKULE KAVWEMI-PHMTG-KT 065
16KIBWABA HILERE PH-MTG-KT120
17KAMWANA SULE PH-MTG-KT 0179
18JONAS KATHALIKO PH-MTG-KT 0304
19KASEREKA LUMANDE ERICK PH-MTG-KT 0110
20 KATHEMBO KAMBILINIYA PH-MTG-KT 069

1.Makinza Watembo.PH-MTG-KIS-NP167
2.Maliani Ami Kambale. PH-MTG-KIS-NP117
3.Kombi Maliyamacho PH-MTG-KIS-NP92
4.Masimengo Alpha Paluku PH-MTH-KIS-NP79
5.Tango Tsika Sarah PH-MTG-NP111
6 Syakaniawiwe Édouard PH-MTG-KIS-NP90
7.Bwanamuzuri Kapunzu Avelu PH-MTG-MA-NP139
8 Yowari Kasereka Albéric PH-MTG-MA-NP49
9.Kairima Thembo Jean Pierre PH-MTG- THA- NP20
10 Kayingi Kitsa Buleli François PH-MTH-THA-NP44

21 KAMBALE WASIKUNDI RINYORE PH-MTG-BU 0302NP
22KASEREKA  VANYAHI JUSTE PH-MTG-BU0307
23KAKULE LONGI HAEI PH-MTG-BU 0324NP
24KAMALE BAKERERI ORIJENE PH-MTGBU0327 NP
25KAMBALE KAMWANA SELE PH-MTG-BU034NP
26KAMBALE MUTHANAVA FIDELE PH-MTG-BU229
27KAMBALE MWENGE ARCENE PH-MTG-BU0322NP
28 KASEREKA NZOMERE JERME PH-MTG-BU0323
29KASEREKA SIVENDAMOLO JABAPTISTE PH-MTG-BU0320
30KASEREKA VATHONDOLI DJUMA PH-MTG-BU289NP
31KAVUWO MUSAYI PH-MTG-BU0314
32MBUSA MWALIKYA ZEPHANIE PH-MTG-BU 166
33MBUSA KAPAMBA PH-MTG-BU 326
34 MUHINDO BAKWANAMAHA JONATHA PH-MTG-BU 303
35MUHINDO MAKUNDO ORIJENE PH-MTG-BU 0324NP
"""

def nettoyer_nom(nom):
    """Nettoie et formate les noms"""
    # Supprime les numéros au début et les caractères spéciaux
    nom = re.sub(r'^\d+[\.:\s]*', '', nom.strip())
    # Nettoie les espaces multiples
    nom = re.sub(r'\s+', ' ', nom)
    return nom.title()

def extraire_codes_et_noms(text):
    """Extrait les noms et codes des données brutes"""
    lignes = text.strip().split('\n')
    noms_sans_code = []
    correspondances = []

    for ligne in lignes:
        ligne = ligne.strip()
        if not ligne:
            continue

        # Cherche les patterns avec codes
        if re.search(r'PH-|CO-|MTG-|MTH-', ligne):
            # Pattern: numéro.nom:code ou numéro nom code
            match = re.match(r'(\d+[\.]?)\s*([^:]+?)\s*(PH-.+|CO-.+|MTG-.+|MTH-.+)', ligne)
            if match:
                nom = nettoyer_nom(match.group(2))
                code = match.group(3).strip()
                correspondances.append((nom, code))
            else:
                # Essai avec séparateur :
                if ':' in ligne:
                    parties = ligne.split(':', 1)
                    if len(parties) == 2:
                        nom_part = re.sub(r'^\d+[\.]?\s*', '', parties[0])
                        nom = nettoyer_nom(nom_part)
                        code = parties[1].strip()
                        correspondances.append((nom, code))
        else:
            # Lignes sans codes (juste des noms)
            nom = nettoyer_nom(ligne)
            if nom and not nom.isspace():
                noms_sans_code.append(nom)

    return correspondances, noms_sans_code

def creer_dataframe_final(correspondances, noms_sans_code):
    """Crée le DataFrame final pour Excel"""
    # Crée une liste pour tous les noms
    tous_les_noms = []

    # Ajoute d'abord les noms avec codes
    noms_avec_codes = [nom for nom, code in correspondances]
    tous_les_noms.extend(noms_avec_codes)

    # Ajoute les noms sans codes (en évitant les doublons)
    for nom in noms_sans_code:
        if nom not in noms_avec_codes:
            tous_les_noms.append(nom)

    # Crée le DataFrame
    data = []
    for nom in tous_les_noms:
        # Trouve le code correspondant
        code_trouve = ""
        for nom_corr, code in correspondances:
            if nom.lower() == nom_corr.lower():
                code_trouve = code
                break

        data.append({
            'Nom du Fermier': nom,
            'Code': code_trouve
        })

    return pd.DataFrame(data)

# Exécution principale
print("Traitement des données en cours...")

# Extraction des données
correspondances, noms_sans_code = extraire_codes_et_noms(donnees_brutes)

print(f"Nombre de correspondances trouvées: {len(correspondances)}")
print(f"Nombre de noms sans codes: {len(noms_sans_code)}")

# Création du DataFrame
df_final = creer_dataframe_final(correspondances, noms_sans_code)

# Sauvegarde en Excel
nom_fichier = "liste_fermiers_avec_codes.xlsx"
df_final.to_excel(nom_fichier, index=False, engine='openpyxl')

print(f"\nFichier Excel créé avec succès: {nom_fichier}")
print(f"Total des fermiers dans le fichier: {len(df_final)}")
print("\nAperçu des données:")
print(df_final.head(10))

# Vérification
print(f"\nVérification:")
print(f"- Noms avec codes: {len(df_final[df_final['Code'] != ''])}")
print(f"- Noms sans codes: {len(df_final[df_final['Code'] == ''])}")
from google.colab import files
files.download("liste_fermiers_avec_codes.xlsx")
```
